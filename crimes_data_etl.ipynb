```python
# Import module pandas and read csv file
import pandas as pd
df = pd.read_csv("crime_data.csv")
```

```python
# Get more information on column types, headers, values
df.info()
```

```python
df.head()
```

```python
# Check to see if there are any null values within the dataset
df.isnull().sum()
```

```python
# Check to see if there are any duplicates - found 1211 duplicate rows
df.duplicated().sum()
```

```python
# Drop duplicates
df.drop_duplicates(inplace=True)
```

```python
df.drop('AREA', axis=1, inplace=True)
```

```python
# Check to see if there any errors within the dates where the date reported is earlier than the date occured
df[df['Date Rptd'] < df['DATE OCC']]
```

```python
# Check to see if there are any erros with the age of the victim being less than 0
df[df['Vict Age'] < 0]
```

```python
# Delete the errored columns that have victim age less than 0
df = df[df['Vict Age'] > 0]
```

```python
# Check if there are genders listed outside of the the scope (M for male, F for female, X for unknown)
gender_list = {'M', 'F', 'X'}
df[-df['Vict Sex'].isin(gender_list)]
```

```python
df = df[df['Vict Sex'].isin(gender_list)]
```

```python
# See if there any ethnicities listed outside of the scope
descent_list = {'A', 'B', 'C', 'D', 'F', 'G', 'H', 'I', 'J', 'K', 'L', 'O', 'P', 'S', 'U', 'V', 'W', 'X', 'Z'}
df[-df['Vict Descent'].isin(descent_list)]
```

```python
# Check to see if there any other statuses than the ones that should be stated
status_list = {'Adult Arrest', 'Adult Other', 'Juv Arrest', 'Juv Other', 'Invest Cont'}
df[-df['Status Desc'].isin(status_list)]
```

```python
df = df[df['Status Desc'].isin(status_list)]
```

```python
df['LOCATION'] = df['LOCATION'].str.strip()
```

```python
#Remove any extra whitespace from the address location
df['LOCATION'].str.replace(r'\s+', ' ', regex=True)
```

```python
df['LOCATION'] = df['LOCATION'].str.replace(r'\s+', ' ', regex=True)
```

```python
count_2020 = df['DATE OCC'][(df['DATE OCC'] >= '2020-01-01') & (df['DATE OCC'] <= '2020-12-31')].count()
count_2021 = df['DATE OCC'][(df['DATE OCC'] >= '2021-01-01') & (df['DATE OCC'] <= '2021-12-31')].count()
count_2022 = df['DATE OCC'][(df['DATE OCC'] >= '2022-01-01') & (df['DATE OCC'] <= '2022-12-31')].count()
count_2023 = df['DATE OCC'][(df['DATE OCC'] >= '2023-01-01') & (df['DATE OCC'] <= '2023-12-31')].count()

print('Number of crimes in 2020:', count_2020, '\nNumber of crimes in 2021:', count_2021, '\nNumber of crimes in 2022:', count_2022, '\nNumber of crimes in 2023:', count_2023)
```

```python
count_2020 = df['Date Rptd'][(df['Date Rptd'] >= '2020-01-01') & (df['Date Rptd'] <= '2020-12-31')].count()
count_2021 = df['Date Rptd'][(df['Date Rptd'] >= '2021-01-01') & (df['Date Rptd'] <= '2021-12-31')].count()
count_2022 = df['Date Rptd'][(df['Date Rptd'] >= '2022-01-01') & (df['Date Rptd'] <= '2022-12-31')].count()
count_2023 = df['Date Rptd'][(df['Date Rptd'] >= '2023-01-01') & (df['Date Rptd'] <= '2023-12-31')].count()

print('Number of crimes in 2020:', count_2020, '\nNumber of crimes in 2021:', count_2021, '\nNumber of crimes in 2022:', count_2022, '\nNumber of crimes in 2023:', count_2023)
```

```python
# We want to focus on crimes that have specifically occured in the years 2021, 2022, and 2023. Therefore, the crimes that occurred in 2020 and were later reported in 2021 will be removed.
df = df[df['DATE OCC'] >= '2021-01-21']
```
